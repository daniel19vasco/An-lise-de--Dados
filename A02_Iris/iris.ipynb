```python
# Célula 01 - Carregar o Dataset Iris e visualizar as primeiras linhas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_iris = sns.load_dataset("iris")
df_iris.to_csv("iris.csv")
```

```python
# Célula 02 - Informações básicas sobre o dataframe (info)

df_iris.info()
```

```python
# Célula 03 - Estatísticas descritivas básicas. Dica: é um método do dataframe
df_iris.describe()
```

```python
# Célula 04 - Contagem de quantas flores existem por espécie

df_iris['species'].value_counts()
```

```python
# Célula 05 - Histograma do campo sepal_length

plt.figure(figsize=(6,4))

# plt.hist(df_iris['sepal_length'], bins=20)
sns.histplot(df_iris['sepal_length'], bins=20)
plt.title("Distribuição de sepal_lenght")
plt.xlabel('sepal_length')
plt.ylabel("Frequência")
plt.show()
```

```python
# Célula 06 - Boxplot de petal_length por espécie

plt.figure(figsize=(6,4))

sns.boxplot(data=df_iris, x='species', y='petal_length')
plt.title("petal_length por espécie")
plt.xlabel("Espécie")
plt.ylabel("petal_length")

plt.show()
```

```python
# Célula 07 - Relação entre sepal_length e sepal_width
# Nesse caso iremos usar o gráfico scatterplot

plt.figure(figsize=(8,4))

sns.scatterplot(
    data=df_iris,
    x='sepal_length',
    y='sepal_width',
    hue='species'
)

plt.title("Relação entre sepal_length e sepal_width")
plt.show()
```

```python
# Célula 08 - Calcule a média de sepal_witdth para cada espécie
# Média é mean()

df_iris.groupby('species')['sepal_width'].mean()
```

```python
# Célula 09 - Calcular quantas flores da espécie *setosa* tem petal_length menor que 1.5

# Dentro dos colchetes do dataframe podemos colocar uma operação com operadores booleanos.
# Será retornado um novo dataframe para cada linha que retorne 'True' para a condição.
setosa = df_iris[df_iris['species'] == 'setosa']
setosa_pequena = setosa[setosa['petal_length'] < 1.5]

print(f"Quantidade de flores da espécie 'setosa' com o comprimento da pétala menor que 1,5 cm: {len(setosa_pequena)}")
```
